# Exploratory Data Analysis (EDA)

## 1. Carga de datos
- Importar el dataset limpio `data/processed/ventas_clean.csv`.

## 2. Estadísticas descriptivas
- Conteo de registros y columnas.
- Promedio, mínimo y máximo de ventas.
- Top productos y clientes.

## 3. Visualizaciones básicas
- Serie temporal de ventas por fecha.
- Barras comparando productos más vendidos.
- Heatmap de ventas por región/mes.

## 4. Insights iniciales
- Productos con mejor desempeño.
- Variación de ventas por periodo.
- Tendencias para planeación de inventario.


```python
# 1. Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Cargar dataset limpio
df = pd.read_csv("data/processed/ventas_clean.csv")

# 3. Estadísticas descriptivas
print("Dimensiones:", df.shape)
print(df.describe(include="all"))

# Top productos
if "producto_id" in df.columns:
    top_productos = df.groupby("producto_id")["cantidad"].sum().sort_values(ascending=False).head(10)
    print("Top productos:\n", top_productos)

# 4. Visualización temporal
if "fecha" in df.columns:
    df.groupby("fecha")["cantidad"].sum().plot(figsize=(10,5))
    plt.title("Ventas por fecha")
    plt.xlabel("Fecha")
    plt.ylabel("Cantidad vendida")
    plt.savefig("visuals/ventas_por_fecha.png")
    plt.show()

# 5. Barras de top productos
if "producto_id" in df.columns:
    sns.barplot(x=top_productos.index, y=top_productos.values)
    plt.title("Top 10 productos más vendidos")
    plt.xticks(rotation=45)
    plt.savefig("visuals/top_productos.png")
    plt.show()
```
