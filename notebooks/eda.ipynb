Análisis avanzado
1. Regresión lineal simple (tendencias de ventas)

Librerias y Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/ventas_clean.csv")  # ajusta ruta según ubicación del notebook
print(df.shape)
df.head()
```

Regresión lineal simple (Tendencia de Ventas)

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Convertir fechas correctamente
df["order_date"] = pd.to_datetime(df["order_date"], dayfirst=True)

# Convertir fecha a número ordinal
df["order_date_num"] = df["order_date"].map(lambda x: x.toordinal())

X = df[["order_date_num"]]
y = df["sales"]

# Ajustar modelo
model = LinearRegression()
model.fit(X, y)

df["pred"] = model.predict(X)

# Visualización
plt.figure(figsize=(10,5))
plt.scatter(df["order_date"], y, label="Ventas reales", alpha=0.5)
plt.plot(df["order_date"], df["pred"], color="red", label="Tendencia (Regresión)")
plt.title("Tendencia de ventas con regresión lineal")
plt.legend()
plt.show()



```

Segmentación de productos (ABC Analysis)

```python
# Calcular ventas totales por producto
ventas_por_producto = df.groupby("product_id")["sales"].sum().sort_values(ascending=False)

# Porcentaje acumulado
ventas_por_producto_pct = ventas_por_producto / ventas_por_producto.sum()
ventas_por_producto_acum = ventas_por_producto_pct.cumsum()

# Clasificación ABC
def clasificar_abc(x):
    if x <= 0.8:
        return "A"
    elif x <= 0.95:
        return "B"
    else:
        return "C"

abc = ventas_por_producto_acum.apply(clasificar_abc)

# Visualización
plt.figure(figsize=(8,5))
ventas_por_producto_acum.plot()
plt.axhline(0.8, color="red", linestyle="--", label="Límite A")
plt.axhline(0.95, color="orange", linestyle="--", label="Límite B")
plt.title("ABC Analysis de productos")
plt.legend()
plt.show()

print("Clasificación ABC:\n", abc.value_counts())
```

Estacionalidad (ventas por mes)

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["mes"] = df["order_date"].dt.month

ventas_por_mes = df.groupby("mes")["sales"].sum()

plt.figure(figsize=(8,5))
ventas_por_mes.plot(kind="bar")
plt.title("Ventas por mes (estacionalidad)")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.show()
```
